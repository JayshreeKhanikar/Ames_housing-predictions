```python
#replacing NA in categorical column with the most frequent value in that column.
def replace_category_NA(df, column):
    try:
        df[column].replace(['NA'], [df[column].value_counts().idxmax()], inplace= True)
    except:
        pass

replace_category_NA(house, 'Electrical') 
replace_category_NA(house, 'MasVnrType')
house['MasVnrType'].value_counts()
# house['Electrical'].value_counts().idxmax()
```

```python
#replacing NA in numerical with median value for that feature.
def replace_numeric_NA(df, column):
    try:
        df[column] = df[column].astype('float64')
    except:
        df[column] = df[column].mask(df[column].str.contains('NA'), np.nan).astype('float64')
        
    df[column] = df[column].mask(df[column].isnull(), df[column].median())
    return

replace_numeric_NA(house,'MasVnrArea')

#house['MasVnrArea'].value_counts()
```

Observing the distribution of SalePrice. 

```python
fig= plt.figure(figsize=(15,10))
sns.distplot(house['SalePrice'])
```

```python
#log transformation of SalePrice to normalize and remove skew from data
house['SalePrice'] = np.log1p(house['SalePrice'])
#house['SalePrice'] = boxcox_normmax(house['SalePrice'], brack=(-1.9, 1.99),  method='mle')
```

```python
#calculating the percentiles and IQR
Q1, Q2, Q3 = np.percentile(house['SalePrice'], 25), np.percentile(house['SalePrice'], 50), np.percentile(house['SalePrice'], 75)
print(Q1, Q2, Q3)
IQR = Q3-Q1
print(IQR)
```

```python
#I am incresing the threshold for Tukey rule on the right side of data to exclude less outliers
o1, o2 = (Q1 - 1.5 * IQR), (Q3 + 2 * IQR)
print(o1,o2)

outliers2 = house['SalePrice'][house['SalePrice'] > o2]
outliers1 = house['SalePrice'][house['SalePrice'] < (Q1 - 1.5 * IQR)]
outliers2 
outliers2.count()
```

```python
#showing SalePrice distr
fig = plt.figure(figsize=(15,8))
_ = sns.distplot(house['SalePrice'])
plt.axvline(Q2, label ="0.5 percentile", c= 'red')
plt.axvline(Q1, label ="0.25 percentile", c= 'orange')
plt.axvline(Q3, linestyle= 'dotted',label ="0.75 percentile", c= 'orange')
plt.axvline((Q1 - 1.5 * IQR),linestyle ='dotted', label ="Tukey lower limit", c= 'green')
plt.axvline((Q3 + 2 * IQR), label ="Tukey upper limit", c= 'lightgreen')
plt.legend()
```

```python
house = house[house['SalePrice'] <(Q3 + 2 * IQR)]
house.shape
```
