```python
#replacing NA in categorical column with the most frequent value in that column.
def replace_category_NA(df, column):
    try:
        df[column].replace(['NA'], [df[column].value_counts().idxmax()], inplace= True)
    except:
        pass

replace_category_NA(house, 'Electrical') 
replace_category_NA(house, 'MasVnrType')
house['MasVnrType'].value_counts()
# house['Electrical'].value_counts().idxmax()
```

```python
#replacing NA in numerical with median value for that feature.
def replace_numeric_NA(df, column):
    try:
        df[column] = df[column].astype('float64')
    except:
        df[column] = df[column].mask(df[column].str.contains('NA'), np.nan).astype('float64')
        
    df[column] = df[column].mask(df[column].isnull(), df[column].median())
    return

replace_numeric_NA(house,'MasVnrArea')

#house['MasVnrArea'].value_counts()
```

```python
#I am incresing the threshold for Tukey rule on the right side of data to exclude less outliers
o1, o2 = (Q1 - 1.5 * IQR), (Q3 + 2 * IQR)
print(o1,o2)

outliers2 = house['SalePrice'][house['SalePrice'] > o2]
outliers1 = house['SalePrice'][house['SalePrice'] < (Q1 - 1.5 * IQR)]
outliers2 
outliers2.count()
```

#### Scaled Benchmark

- Fit the `benchmark_scaled` model against the scaled training data. 
- Score the `benchmark_scaled` model against both the scaled training set and the scaled testing set. 
- Write the results to the results `list`.

## Explain the bias / variance tradeoff

Why do regularized models perform better on your test set?

Regularized models impose either l1 or l2 penalty on the features in order to prevent over fitting the model. This wil result in little bit more variance on training dataset but help to decrease bias on test data.

### Interpreting coefficients

For your best model, 

* plot relevant coefficients using the `plot_coef` functoin.


```python
def plot_coef(model, top_n = 10):
    '''
    Plots the magnitude of top and bottom n coefficients
    '''
    cols = house_train.columns
    coef = model.coef_
    zipped = list(zip(cols, coef))
    zipped.sort(key=lambda x: x[1], reverse = True)
    top_10 = pd.DataFrame(zipped).head(top_n)
    bottom_10 = pd.DataFrame(zipped).tail(top_n)
    return pd.concat([top_10, bottom_10], axis=0).plot.barh(x = 0, y = 1)
```
