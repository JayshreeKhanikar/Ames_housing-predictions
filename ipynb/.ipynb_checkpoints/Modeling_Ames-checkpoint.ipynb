## Modelling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

from IPython.display import display 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import  mean_squared_error
```

```python
ames_feature = pd.read_pickle('../data/ames_feature.p')
target = pd.read_pickle('../data/target.p')
```

```python
fig= plt.figure(figsize=(10,10))
sns.distplot(target)
```

```python
#pca_ames.explained_variance_
```

## Split your data into train and test set

```python
house_train, house_test, y_train, y_test = train_test_split(ames_feature, target, test_size= 0.5)
house_train.shape, house_test.shape
```

## Scale the data

```python
numerical_feat = house_train.select_dtypes(include=['int', 'float64']).columns
```

```python
sk = StandardScaler()
sk_fit = sk.fit(house_train.loc[:,numerical_feat])
house_train.loc[:,numerical_feat] = sk_fit.transform(house_train.loc[:,numerical_feat])
house_test.loc[:,numerical_feat] = sk_fit.transform(house_test.loc[:,numerical_feat])
```

```python
lr = LinearRegression()
lr.fit(house_train, y_train)
lr.score(house_train, y_train), lr.score(house_test, y_test)# R2 score
```

```python
mean_squared_error(y_test, lr.predict(house_test))
```

```python
pipe_lr = Pipeline([('pca', PCA(150)),
                   ('lr', LinearRegression())])
pipe_lr.fit(house_train, y_train)
pipe_lr.score(house_train, y_train), pipe_lr.score(house_test, y_test)# R2 score
```

```python
#pipe_lr.steps[1][1].coef_
pipe_lr.steps[0][1].components_.shape
```

```python
mean_squared_error(y_test, pipe_lr.predict(house_test))
```

Unregularized linear model is really worse. I think there are lots of features that have no predictive value and or are dependent on other features. I tried to reduce the features with PCA and then used the LinearRegression, which improved my model. Regularization will also take care of this. 

## Regularized model: Lasso

```python
pipe_lasso = Pipeline([('lasso', Lasso())])
lasso_param = {'lasso__alpha': np.logspace(-4, -1, 40)}

gr_lasso = GridSearchCV(pipe_lasso, param_grid=lasso_param, cv= 5, n_jobs=-1)
```

```python
gr_lasso.fit(house_train, y_train)
```

```python
gr_lasso.score(house_train, y_train)
```

```python
gr_lasso.score(house_test,y_test)
```

```python
#gr_lasso.grid_scores_
```

```python
#getting coef from Lasso estimator
#gr_lasso.best_estimator_.steps[2][1].coef_
```

```python
gr_lasso.best_score_
```

```python
gr_lasso.best_params_
```

```python
predictions = gr_lasso.best_estimator_.steps[0][1].predict(house_test)
```

```python
mean_squared_error(y_test, gr_lasso.best_estimator_.steps[0][1].predict(house_test))
```

```python
fig= plt.figure(figsize=(8,8))
plt.scatter(y_test, gr_lasso.predict(house_test), c ='r', label= 'test')
plt.scatter(y_train, gr_lasso.predict(house_train), c= 'lightgreen', label='train')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.legend()
```

```python
fig= plt.figure(figsize=(8,8))
plt.scatter(gr_lasso.predict(house_test), y_test-gr_lasso.predict(house_test), c='r', label='test')
plt.scatter(gr_lasso.predict(house_train), y_train-gr_lasso.predict(house_train), c='lightgreen', label='train')
plt.xlabel('Predicted Sale Price')
plt.ylabel('Residuals')
plt.legend()
```

```python
lasso_feat=pd.DataFrame(gr_lasso.best_estimator_.steps[0][1].coef_, columns=['beta'], index= house_test.columns).\
                         sort_values('beta', ascending= False)
pd.concat([lasso_feat.head(10), lasso_feat.tail(10)]).plot(kind='bar')
```

```python
len(house_test.columns[gr_lasso.best_estimator_.steps[0][1].coef_ ==0])
```

##### Lasso made the coefficients of 176 as 0 and only kept 78 features as informative. Top features driving the SalePrice of house are GrLivArea, Neighborhood and Condition1. Features that degrade the house values are LandContour, bad sale condition and Claytile Roof Material.

## Regularized model: ElasticNet

```python
pipe_en = Pipeline([('en', ElasticNet())])
en_param = {'en__alpha': np.logspace(-3, -1, 20),
           'en__l1_ratio': np.linspace(0,1,10) }
en_gr = GridSearchCV(pipe_en, param_grid=en_param, cv=5, n_jobs=-1)
```

```python
en_gr.fit(house_train, y_train)
```

```python
en_gr.score(house_train, y_train), en_gr.score(house_test, y_test)
```

```python
en_gr.best_params_
```

```python
mean_squared_error(y_test, en_gr.predict(house_test))
```

```python
fig= plt.figure(figsize=(8,8))
plt.scatter(y_test, en_gr.predict(house_test), c='r', label='test')
plt.scatter(y_train, en_gr.predict(house_train), c='lightblue', label='train')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.legend()
```

```python
fig= plt.figure(figsize=(8,8))
plt.scatter(en_gr.predict(house_test), y_test-en_gr.predict(house_test), c='r', label='test')
plt.scatter(en_gr.predict(house_train), y_train-en_gr.predict(house_train), c='lightblue', label='train')
plt.xlabel('Predicted Sale Price')
plt.ylabel('Residuals')
plt.legend()
```

```python
en_feat = pd.DataFrame(en_gr.best_estimator_.steps[0][1].coef_, columns=['beta'], index=house_test.columns).\
            sort_values('beta', ascending= False)
pd.concat([en_feat.head(10),en_feat.tail(10)]).plot(kind='bar')
```

```python
len(house_test.columns[en_gr.best_estimator_.steps[0][1].coef_ ==0])
```

##### Lasso made the coefficients of 171 as 0 and only kept 83 features as informative. Top features driving the SalePrice of house are GrLivArea, Neighborhood and Condition1. Features that degrade the house values are LandContour, bad sale condition and Claytile Roof Material.

## Tree Based model: DecisionTree

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(house_train, y_train)
```

```python
dt.score(house_train, y_train)
```

```python
dt.score(house_test, y_test)
```

```python
mean_squared_error(y_test, dt.predict(house_test))
```

```python
fig = plt.figure(figsize=(8,8))
plt.scatter(y_test, dt.predict(house_test), c='blue', label='test')
plt.scatter(y_train, dt.predict(house_train), c='lightgreen', label='train')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.legend()
```

```python
fig = plt.figure(figsize=(8,8))
plt.scatter(dt.predict(house_test), y_test-dt.predict(house_test), c='blue', label='test')
plt.scatter(dt.predict(house_train), y_train-dt.predict(house_train), c='lightgreen', label='train')
plt.xlabel('Predicted Sale Price')
plt.ylabel('Residual')
plt.legend()
```

We can clearly see that DecisionTree overfits the trainingdata.

```python
importance =(dt.feature_importances_)
len(house_test.columns[importance==0])
```

```python
dt_feat = pd.DataFrame(importance, columns=['Feat_imp'], index=house_test.columns).sort_values('Feat_imp', ascending=False)
```

```python
pd.concat([dt_feat.head(10),dt_feat.tail(10)]).plot(kind='bar')
```

##### Decisiontree made the coefficients of 123 as 0 and only kept 131 features as informative. Top features driving the SalePrice of house are OverallQual, GrLivArea and TotalBsmtSF. Features that degrade the house values are bad Exterior Material.

## Conclusion:
Most importance features that drive the House Prise in Ames are OverallQuality, GrLivArea and Neighborhood. From the EDA work done before it was clear the GrLivArea and OverallQual strong corelation with SalePrice. Though Neighborhood did not come up in our EDA, it makes sense that our models picked it up as real estate value is driven by location.
